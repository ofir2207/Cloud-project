```python
# @title Student Management System
# @markdown Fill in the details and click the button to update the file.

import ipywidgets as widgets
from IPython.display import display
from google.colab import drive

# Mount Google Drive

# Path to the file
file_path = '/content/drive/MyDrive/students.txt'

# Function to load data from the TXT file
def load_students(path):
    data = {}
    try:
        with open(path, 'r', encoding='utf-8') as f:
            for line in f:
                parts = [p.strip() for p in line.strip().split('|')]
                if len(parts) >= 4:
                    # Name, Last Name, Email, Courses, Link
                    data[parts[0]] = {
                        'email': parts[1],
                        'courses': parts[2],
                        'link': parts[3]
                    }
    except FileNotFoundError:
        print("Error: students.txt not found. Please check the path.")
    return data

students_dict = load_students(file_path)

# Create UI Widgets
name_select = widgets.Dropdown(options=['Select Student'] + list(students_dict.keys()), description='Name:')
email_box = widgets.Text(description='Email:', disabled=True)
courses_box = widgets.Text(description='Courses:', disabled=True)
link_box = widgets.Text(description='Link:', disabled=True)
favorite_show = widgets.Text(description='Fav Show:', placeholder='Enter show name...')
update_btn = widgets.Button(description="Update Student", button_style='info') # Blue button
output = widgets.Output()

# Update text boxes when a name is selected
def on_name_change(change):
    student_name = change['new']
    if student_name in students_dict:
        info = students_dict[student_name]
        email_box.value = info['email']
        courses_box.value = info['courses']
        link_box.value = info['link']
    else:
        email_box.value = courses_box.value = link_box.value = ""

name_select.observe(on_name_change, names='value')

# Handle Button Click (Append to file)
def on_button_clicked(b):
    with output:
        output.clear_output()
        if name_select.value == 'Select Student' or not favorite_show.value:
            print("Please select a student and enter a show name.")
            return

        new_entry = f"\n{name_select.value}, {favorite_show.value}"
        with open(file_path, 'a', encoding='utf-8') as f:
            f.write(new_entry)
        print(f"Success! '{favorite_show.value}' added for {name_select.value}.")

update_btn.on_click(on_button_clicked)

# Display the Form
# Organizing the layout into two columns for a "Form" look
form_items = [
    name_select,
    widgets.HBox([email_box, courses_box]),
    widgets.HBox([link_box, favorite_show]),
    update_btn,
    output
]
form = widgets.VBox(form_items)
display(form)
```
