```python
from firebase import firebase
import time

FBconn = firebase.FirebaseApplication('https://tar3room7-4e48a-default-rtdb.firebaseio.com/', None)
def get_all_records():
    return FBconn.get('/myTest1/', None)

def display_records():
    records = get_all_records()
    if records:
        print("\nCurrent Records:")
        for record_id, data in records.items():
            print(f"ID: {record_id} - Word: {data['word']} Count: {data['count']}")
    else:
        print("\nNo records found")


def sync_word(word_to_find):
    # 1. Get all data from the collection
    result = FBconn.get('/myTest1/', None)

    if result:
        # 2. Look for the word in the records
        for record_id, data in result.items():
            if data.get('word') == word_to_find:
                # 3. If found, update the count
                new_count = data.get('count', 0) + 1
                FBconn.put(f'/myTest1/{record_id}', 'count', new_count)
                print(f"Updated '{word_to_find}' count to {new_count}")
                return True

    # 4. If not found, add it as a new entry
    new_data = {'word': word_to_find, 'count': 1}
    FBconn.post('/myTest1/', new_data)
    print(f"Added new word: '{word_to_find}'")
    return False

while True:
    print("\nCommon words counter menu:")
    print("1. Add a single word")
    print("2. Add text for analysis")
    print("3. Update word count")
    print("4. Delete word")
    print("5. View all words")
    print("6. Exit")

    choice = input("\nEnter your choice (1-6): ")

    if choice == '1':
        word = input("What is the word? ")
        sync_word(word)
    elif choice == '2':
        sentence = input("What is the text? ")
        words = sentence.split(" ")
        for word in words:
          sync_word(word)

    elif choice == '3':
        display_records()
        try:
          record_id = input("\nEnter the record ID to update: ")
          new_count = int(input("Enter the new count: "))
          if new_count>0:
            FBconn.put('/myTest1/' + record_id, 'count', new_count)
          else:
            print("Please enter a positive number for the count")
        except ValueError:
          print("Please enter a valid number for the count")


    elif choice == '4':
        # View all records
        display_records()
        record_id = input("\nEnter the record ID to delete: ")
        FBconn.delete('/myTest1/',record_id)

    elif choice == '5':
      display_records()

    elif choice== '6':
      break

    else:
        print("Invalid choice. Please try again.")
```
