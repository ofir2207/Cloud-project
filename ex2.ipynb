```python
# @title Student Management System

import json
import os

# Path to your JSON file
file_path = '/content/drive/MyDrive/students.json'

def get_student_names(path):
    if os.path.exists(path):
        with open(path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            # Data is a list of dictionaries, extract 'name' from each
            return [student['name'] for student in data if 'name' in student]
    return ["No students found"]

# --- SETUP THE FORM ---
# Use the actual names from your JSON here in the brackets to see them in the dropdown
# After you run the cell once, you can update these strings manually if the file changes
student_name = "Ofir Kortovitch" #@param ["Ofir Kortovitch", "Liraz Tubul"]
favorite_show = "friends" #@param {type:"string"}

def update_json(path, name, show):
    if not os.path.exists(path):
        print("Error: JSON file not found.")
        return

    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f) # data is now a list of dictionaries

    found = False
    for student in data:
        if student.get('name') == name:
            student['favorite_show'] = show
            found = True
            break

    if found:
        with open(path, 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=4, ensure_ascii=False)
        print(f"Success: Added '{show}' to {name}")
    else:
        print(f"Error: {name} not found in the student list.")

# Execution
if favorite_show:
    update_json(file_path, student_name, favorite_show)
else:
    print("Please enter a favorite show name to update.")

# Display current list of names found in file (for your reference)
print("\nAvailable names in JSON file:", get_student_names(file_path))
```
